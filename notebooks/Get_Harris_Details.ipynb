```python
import pandas as pd
import json
import http.client
import random
from api_keys import attom_data_key
```

```python
a = pd.read_csv(f"../data/rawdata/Harris_County.csv",header=None)
zips =  list(a[0])
```

```python
conn = http.client.HTTPSConnection("search.onboard-apis.com") 

headers = { 
    'accept': "application/json", 
    'apikey': attom_data_key
} 

total_Addresses = []
county_Zip_Adds = pd.DataFrame(data={"zip":"", "address":"", "lat":"", "lng":""}, index = [0])

for i in range(0,len(zips)):
        print(f"Harris, {zips[i]}")
        for tries in range(1,11):
            try:
                pages = []
                page_Num = 2
                conn.request("GET", f"https://search.onboard-apis.com/propertyapi/v1.0.0/property/address?postalcode={zips[i]}&page=1&pagesize=2000", headers=headers)
                res = conn.getresponse().read().decode()
                data = json.loads(res)
                page_Size = len(data["property"])
                total = data["status"]["total"]
                total_Addresses.append(total)
                print(f"There are {total} number of addresses known in this zipcode.")
                total -= 2000
                while total > 0:
                    pages.append(page_Num)
                    page_Num += 1
                    total -= 2000
                print(f"\tNum Results: {page_Size} for page 1")
                if page_Size > 0:
                    df = pd.DataFrame(data={"zip":"", "address":"", "lat":"", "lng":""}, index = [k for k in range(page_Size)])
                    for j in range(page_Size):
                        df.iat[j, 0] = data['property'][j]['address']['postal1']
                        df.iat[j, 1] = data['property'][j]['address']['oneLine']
                        df.iat[j, 2] = data['property'][j]['location']['latitude']
                        df.iat[j, 3] = data['property'][j]['location']['longitude']
                    df.drop_duplicates(subset="address",inplace=True)
                    df = df.reset_index(drop=True)
                    county_Zip_Adds = county_Zip_Adds.append(df.copy(),ignore_index = True)
                for c in pages:
                    for add_tries in range(1,11):
                        try:
                            conn.request("GET", f"https://search.onboard-apis.com/propertyapi/v1.0.0/property/address?postalcode={zips[i]}&page={c}&pagesize=2000", headers=headers)
                            res = conn.getresponse().read().decode()
                            data = json.loads(res)
                            page_Size = len(data["property"])
                            print(f"\tNum Results: {page_Size} for page {c}")
                            if page_Size > 0:
                                df = pd.DataFrame(data={"zip":"", "address":"", "lat":"", "lng":""}, index = [k for k in range(page_Size)])
                                for j in range(page_Size):
                                    df.iat[j, 0] = data['property'][j]['address']['postal1']
                                    df.iat[j, 1] = data['property'][j]['address']['oneLine']
                                    df.iat[j, 2] = data['property'][j]['location']['latitude']
                                    df.iat[j, 3] = data['property'][j]['location']['longitude']
                                df.drop_duplicates(subset="address",inplace=True)
                                df = df.reset_index(drop=True)
                                county_Zip_Adds = county_Zip_Adds.append(df.copy(),ignore_index = True)
                            break
                        except:
                            print(f"\tTry {add_tries} for page {c} didn't parse correctly. Trying again up to 10 times total.")
                break
            except:
                    print(f"\tTry {tries} for page 1 didn't parse correctly. Trying again up to 10 times total.")
county_Zip_Adds = county_Zip_Adds.drop([0])
county_Zip_Adds = county_Zip_Adds.reset_index(drop=True)
```

```python
group_Adds = pd.DataFrame(data={"zip":"", "address":"", "lat":"", "lng":""}, index = [0])

for i in range(len(total_Addresses)):#len(total_Addresses)
    df = county_Zip_Adds.loc[county_Zip_Adds["zip"] == str(zips[i])]
    df = df.reset_index(drop=True)
    length = len(df.index)
    
    if length >= 500:
        rands = random.sample(range(0,length),500)
        group_Adds = group_Adds.append(df.iloc[rands],ignore_index = True)
        
group_Adds = group_Adds.drop([0])
group_Adds = group_Adds.reset_index(drop=True)
```

```python
group_Adds.to_csv("../data/rawdata/randomly_Sampled_Addresses.csv",index=False)
```
