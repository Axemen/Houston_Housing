```python
# Dependencies
import pandas as pd
```

```python
# Read HPD crime report count data and save as "crime_info_ct"
crime_info_ct =pd.read_csv("../data/cleandata/HPD_Crime_Report_Counts.csv")

# Preview "crime_info_ct"
crime_info_ct.head()
```

```python
# Read Houston Harris County zipcodes data and save as "harris_cty_zipcode_pd"
harris_cty_zipcode_pd = pd.read_csv("../data/cleandata/Harris_County_zipcode_lat_lngs.csv")

# Preview "harris_cty_zipcode_pd"
harris_cty_zipcode_pd.head()
```

```python
# Calculate crime counts in Harris County
harris_cty_crime_ct = pd.merge(crime_info_ct, harris_cty_zipcode_pd, on="ZIP")

# Preview "harris_cty_crime_ct"
harris_cty_crime_ct.head()        
```

```python
# Add new column for total crime count
harris_cty_crime_ct["Total Ct"] = harris_cty_crime_ct["Non-Violent Ct"] + harris_cty_crime_ct["Violent Ct"]

# Preview "harris_cty_crime_ct"
harris_cty_crime_ct.head()        
```

## Crime Counts Calculator

```python
# Write "harris_cty_crime_ct" to csv file
harris_cty_crime_ct.to_csv("../data/cleandata/Harris_County_Crime_Ct_by_zipcode.csv", index=False, header=True)
```
