```python
from api_keys import attom_data_key
import http.client 
import json
import pandas as pd

conn = http.client.HTTPSConnection("search.onboard-apis.com") 

headers = { 
    'accept': "application/json", 
    'apikey': attom_data_key
} 

countList = ["Austin","Brazoria","Chambers","Fort_Bend","Galveston","Harris","Liberty","Montgomery","Waller"]
with open("County_Zips.json","r") as thing:
    zips = json.load(thing)
    print(zips["Austin"])
```

```python
#data['property'][1999] Property info for the 1999th record of the 2000 returned
df = pd.DataFrame(data={"zip":"", "address":"", "lat":"", "lng":""},index=[0])
county_Zip_Adds = {county:df for county in countList}
for county in ["Austin"]:#countList
    for i in range(0,len(zips[county])):#range(0,len(zips[county]))
        print(f"{zips[county][i]}")
        conn.request("GET", f"https://search.onboard-apis.com/propertyapi/v1.0.0/property/address?postalcode={zips[county][i]}&page=1&pagesize=2000", headers=headers)
        res = conn.getresponse().read().decode()
        data = json.loads(res)
        if data["status"]["total"] <= 2000:
            page_Size = data["status"]["total"]
        else:
            page_Size = 2000
        df = pd.DataFrame(data={"zip":"", "address":"", "lat":"", "lng":""}, index = [k for k in range(page_Size)])
        for j in range(page_Size):
            df.iat[j, 0] = data['property'][j]['address']['postal1']
            df.iat[j, 1] = data['property'][j]['address']['oneLine']
            df.iat[j, 2] = data['property'][j]['location']['latitude']
            df.iat[j, 3] = data['property'][j]['location']['longitude']
        df.drop_duplicates(subset="address",inplace=True)
        df = df.reset_index(drop=True)
        county_Zip_Adds[county] = county_Zip_Adds[county].append(df.copy(),ignore_index = True)
    county_Zip_Adds[county] = county_Zip_Adds[county].drop([0])
    county_Zip_Adds[county] = county_Zip_Adds[county].reset_index(drop=True)
county_Zip_Adds["Austin"]
```
