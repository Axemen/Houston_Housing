```python
import pandas as pd
import json
countList = ["Austin","Brazoria","Chambers","Fort_Bend","Galveston","Harris","Liberty","Montgomery","Waller"]
zips = {}
```

```python
for name in countList:
    a = pd.read_csv(f"../data/rawdata/{name}_County.csv",header=None)
    zips[name] = list(a[0])
```

```python
#Open using with open(name, "r") as thing: input = json.load(thing)
with open('County_Zips.json','w') as outs:
    json.dump(zips,outs)
```
